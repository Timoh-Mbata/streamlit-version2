* CREATING A SOCIAL ECONOMIC INDICATOR
* economic Participation Index (EPI):
Components: bank_account, cellphone_access, job_type
- Interpretation: A score reflecting access to financial resources and employment stability.
- Socioeconomic Status Index (SSI):
- Components: education_level, marital_status, gender_of_respondent, location_type
- Interpretation: A score indicating the potential socio-economic standing based on educational background and social factors.
Overall Socio-Economic Status:
- Combine the EPI and SSI to form a comprehensive status score. This can be done through simple addition or weighted averages depending on the perceived importance of each component.

```python
dataset['EPI'] = dataset['bank_account'] + dataset['cellphone_access'] + dataset['job_type'] 
dataset['SSI'] = dataset['education_level'] +  dataset['marital_status'] +  dataset['gender_of_respondent'] +  dataset['location_type']
```

```python
dataset['SOCIAL_ECONOMIC_STATUS'] = dataset['EPI'] + dataset['SSI']
```

```python
dataset.sample(5)
```

```python
age_groups = ['child','youth','Adult','Elderly']
```

```python
for age in dataset['age_of_respondent'] :
    if age <= 18:
        dataset['age_group'] = 0
    elif (age > 19) & (age <= 34):
        dataset['age_group'] = 1
    elif (age >35 ) & (age <= 63):
        dataset['age_group'] = 2
    else :
        dataset['age_group'] = 3
```

# MODEL BUIDING USING LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x = dataset2.drop(columns='bank_account')
y = dataset2.bank_account
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=100)
```

```python
model = LogisticRegression(random_state=42)
```

```python
X_train.columns
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
